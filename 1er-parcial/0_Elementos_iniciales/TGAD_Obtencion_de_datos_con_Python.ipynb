Tenga en cuenta que solo es posible realizar hasta 100 consultas por día.

Paso 4:

```python
#importar librerías
import requests
import pandas as pd
```

```python
#seteo de token
import os
from dotenv import load_dotenv
load_dotenv()
auth_token = os.getenv('BCRA_TOKEN') #Token del BCRA desde archivo .env
head = {'Authorization': 'Bearer ' + auth_token}
```

```python
#Por ejemplo, se consulta la serie de cotización del dólar
url = 'https://api.estadisticasbcra.com/usd'
response = requests.get(url, headers=head)
```

```python
#se obtiene la serie en formato json
json_data=response.json()
```

```python
#se convierte a DataFrame
df=pd.DataFrame(json_data)
df
#df.sort_values(by="d",ascending=False) opcional
```

```python
#se agrega nombre a las columnas
df.columns=['Fecha', 'Cotización del Dolar']
df
```
