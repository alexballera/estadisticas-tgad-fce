```python
def TH_DIFMEDIA_VARCON(MM_1, MM_2, DIF, sigma2_1, sigma2_2, n_1, n_2, alfa, simbolo_H1):
  import scipy.stats as ss
  import numpy as np
  Zobs=round(((MM_1-MM_2)-DIF) / (np.sqrt((sigma2_1/n_1)+(sigma2_2/n_2))),4)
  if simbolo_H1=='>':
    Zalfa  = round(ss.norm.ppf(1-alfa, 0,1),4)
    if Zobs > Zalfa:
      respuesta='Como se cumple C.D ya que ' + 'Zobs='+str(Zobs)+' es mayor que '+ 'Z_1-alfa='+str(Zalfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + 'Zobs='+str(Zobs)+' es menor que '+ 'Z_1-alfa='+str(Zalfa)+' entonces no se rechaza H0'

    p_valor=1-ss.norm.cdf(Zobs, 0,1)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  elif simbolo_H1=='<':
    Zalfa  = round(ss.norm.ppf(alfa, 0,1),4)
    if Zobs < Zalfa:
      respuesta='Como se cumple C.D ya que ' + 'Zobs='+str(Zobs)+' es menor que '+ 'Zalfa='+str(Zalfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + 'Zobs='+str(Zobs)+' es mayor que '+ 'Zalfa='+str(Zalfa)+' entonces no se rechaza H0'

    p_valor=ss.norm.cdf(Zobs, 0,1)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  else:
    Zalfa  = round(ss.norm.ppf(1-alfa/2, 0,1),4)
    if np.abs(Zobs) > Zalfa:
      respuesta='Como se cumple C.D ya que ' + '|Zobs|='+'|'+str(Zobs)+'|'+' es mayor que '+ 'Z_1-alfa/2='+str(Zalfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + '|Zobs|='+'|'+str(Zobs)+'|'+' es menor que '+ 'Z_1-alfa/2='+str(Zalfa)+' entonces no se rechaza H0'

    p_valor=2*(1-ss.norm.cdf(abs(Zobs), 0,1))
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  return respuesta, respuesta1
```

## Enunciado

Con el fin de estudiar la efectividad de un nuevo aditivo para combustible se realizÃ³ la 
siguiente prueba: se tomaron 20 autos de la misma marca y modelo, a 10 de ellos se les colocÃ³ 
el aditivo y los otros 10 quedaron como control. Luego de 10 corridas sobre el mismo trayecto, 
se obtuvo para los autos con el aditivo una media de 4,2 litros con un desvÃ­o estÃ¡ndar de 0,58 
litros. Los autos control dieron un consumo medio de 5,25 litros con un desvÃ­o de 0,63 litros. 
Poner a prueba la hipÃ³tesis de igualdad de varianzas con un nivel de significaciÃ³n del 5%.

## Datos del problema

**Grupo con aditivo:**
- TamaÃ±o de muestra: $n_1 = 10$
- Media muestral: $\bar{x}_1 = 4.2$ litros
- DesvÃ­o estÃ¡ndar muestral: $s_1 = 0.58$ litros
- Varianza muestral: $s_1^2 = 0.3364$ litrosÂ²

**Grupo control (sin aditivo):**
- TamaÃ±o de muestra: $n_2 = 10$
- Media muestral: $\bar{x}_2 = 5.25$ litros
- DesvÃ­o estÃ¡ndar muestral: $s_2 = 0.63$ litros
- Varianza muestral: $s_2^2 = 0.3969$ litrosÂ²

**Nivel de significaciÃ³n:** $\alpha = 0.05$

**Objetivo:** Poner a prueba la hipÃ³tesis de igualdad de varianzas (test bilateral).

### FunciÃ³n auxiliar para Test de Igualdad de Varianzas

```python
def TH_IGUALDAD_VARIANZA(var_M1, var_M2, n_1, n_2, alfa, simbolo_H1):
    import scipy.stats as ss
    import numpy as np
    Fobs = round(var_M1 / var_M2, 4)
    if simbolo_H1 == '>':
        falfa = round(ss.f.ppf(1-alfa, n_1-1, n_2-1), 4)
        if Fobs > falfa:
            respuesta = 'Como se cumple C.D ya que ' + 'Fobs=' + str(Fobs) + ' es mayor que ' + 'F_1-alfa;n1-1;n2-1=' + str(falfa) + ' entonces se rechaza H0'
        else:
            respuesta = 'Como no se cumple C.D ya que ' + 'Fobs=' + str(Fobs) + ' es menor que ' + 'F_1-alfa;n1-1;n2-1=' + str(falfa) + ' entonces no se rechaza H0'
        
        p_valor = round(1 - ss.f.cdf(Fobs, n_1-1, n_2-1), 8)
        if p_valor < alfa:
            respuesta1 = 'Como se cumple C.D ya que ' + 'P-valor=' + str(p_valor) + ' es menor que ' + 'alfa=' + str(alfa) + ' entonces se rechaza H0'
        else:
            respuesta1 = 'Como no se cumple C.D ya que ' + 'P-valor=' + str(p_valor) + ' es mayor que ' + 'alfa=' + str(alfa) + ' entonces no se rechaza H0'
    
    elif simbolo_H1 == '<':
        falfa = round(ss.f.ppf(alfa, n_1-1, n_2-1), 4)
        if Fobs < falfa:
            respuesta = 'Como se cumple C.D ya que ' + 'Fobs=' + str(Fobs) + ' es menor que ' + 'F_alfa;n1-1;n2-1=' + str(falfa) + ' entonces se rechaza H0'
        else:
            respuesta = 'Como no se cumple C.D ya que ' + 'Fobs=' + str(Fobs) + ' es mayor que ' + 'F_alfa;n1-1;n2-1=' + str(falfa) + ' entonces no se rechaza H0'
        
        p_valor = round(ss.f.cdf(Fobs, n_1-1, n_2-1), 8)
        if p_valor < alfa:
            respuesta1 = 'Como se cumple C.D ya que ' + 'P-valor=' + str(p_valor) + ' es menor que ' + 'alfa=' + str(alfa) + ' entonces se rechaza H0'
        else:
            respuesta1 = 'Como no se cumple C.D ya que ' + 'P-valor=' + str(p_valor) + ' es mayor que ' + 'alfa=' + str(alfa) + ' entonces no se rechaza H0'
    
    else:
        falfa = round(ss.f.ppf(1-alfa/2, n_1-1, n_2-1), 4)
        if np.abs(Fobs) > falfa:
            respuesta = 'Como se cumple C.D ya que ' + '|Fobs|=' + '|' + str(Fobs) + '|' + ' es mayor que ' + 'F_1-alfa/2;n1-1;n2-1=' + str(falfa) + ' entonces se rechaza H0'
        else:
            respuesta = 'Como no se cumple C.D ya que ' + '|Fobs|=' + '|' + str(Fobs) + '|' + ' es menor que ' + 'F_1-alfa/2;n1-1;n2-1=' + str(falfa) + ' entonces no se rechaza H0'
        
        p_valor = round(2 * min(ss.f.cdf(Fobs, n_1-1, n_2-1), 1 - ss.f.cdf(Fobs, n_1-1, n_2-1)), 8)
        if p_valor < alfa:
            respuesta1 = 'Como se cumple C.D ya que ' + 'P-valor=' + str(p_valor) + ' es menor que ' + 'alfa=' + str(alfa) + ' entonces se rechaza H0'
        else:
            respuesta1 = 'Como no se cumple C.D ya que ' + 'P-valor=' + str(p_valor) + ' es mayor que ' + 'alfa=' + str(alfa) + ' entonces no se rechaza H0'
    
    return respuesta, respuesta1
```

### ResoluciÃ³n siguiendo los 8 pasos

```python
# Paso 1: Definir las variables aleatorias
display(Math(r'1) \thinspace X_1=\text{"consumo de combustible con aditivo"}'))
print('')
display(Math(r'\thinspace \thinspace \thinspace X_2=\text{"consumo de combustible sin aditivo (control)"}'))
print('')

# Paso 2: Distribuciones y condiciones
display(Math(r'2) \thinspace X_1 \sim N(\mu_1; \sigma_1) \thinspace X_2 \sim N(\mu_2; \sigma_2)'))
print('')

# Paso 3: HipÃ³tesis
display(Math(r'3) \thinspace H_0: \sigma_1^2 = \sigma_2^2 \thinspace \thinspace H_1: \sigma_1^2 \neq \sigma_2^2'))
print('')

# Pasos 4 y 5: EstadÃ­stico de prueba y su distribuciÃ³n
display(Math(r'4 \thinspace y \thinspace 5) \thinspace F_{obs}= {S_1^2 \over S_2^2} \sim F_{n_1-1;n_2-1}'))
print('')

# Paso 6: Criterio de decisiÃ³n
display(Math(r'6) \thinspace CD: |F_{obs}| > F_{1- {\alpha \over 2};n_1-1;n_2-1} \thinspace para \thinspace rechazar \thinspace H_0'))
print('')

# Paso 7: Aplicar el test
display(Math(r'7)'))
# Datos
s1_sq = 0.58**2  # Varianza muestral grupo con aditivo
s2_sq = 0.63**2  # Varianza muestral grupo control
n1 = 10
n2 = 10
alpha = 0.05

# Ejecutar el test
print(TH_IGUALDAD_VARIANZA(s1_sq, s2_sq, n1, n2, alpha, '')[0])
print(TH_IGUALDAD_VARIANZA(s1_sq, s2_sq, n1, n2, alpha, '')[1])
print('')

# Paso 8: ConclusiÃ³n contextualizada
print('8) No hay evidencia estadÃ­sticamente significativa con un 95% de confianza para')
print('   concluir que las varianzas del consumo de combustible son diferentes entre')
print('   el grupo con aditivo y el grupo control. Por lo tanto, es vÃ¡lido asumir')
print('   igualdad de varianzas para anÃ¡lisis posteriores.')
```

### VisualizaciÃ³n del test (Adicional)
