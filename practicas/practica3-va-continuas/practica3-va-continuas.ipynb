# Ejercicio 1

**Una variable aleatoria continua X toma valores sólo entre 0 y 4 y su función de densidad está dada por f(x) = 1/4 en ese intervalo.**

a) Calcular:

1. $P(X < 2)$
2. $P(X \geq 4)$
3. $P(0 < X < 3)$
4. $P(X > 3)$

b) Calcular

$P(X < 3 | X > 2)$ y $P(X > 1 | X < 3.5)$

c) Hallar y graficar la función de distribución acumulada

## Solución

### Identificación de la distribución

La variable aleatoria **X** sigue una **distribución uniforme continua** en el intervalo [0, 4]:

$$X \sim U(0, 4)$$

**Función de densidad:**

$$f(x) = \begin{cases} 
\frac{1}{4} & \text{si } 0 \leq x \leq 4 \\
0 & \text{en otro caso}
\end{cases}$$

**Parámetros:**
- $a = 0$ (límite inferior)
- $b = 4$ (límite superior)

```python
# Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from IPython.display import display, Math
import sympy as sp
```

### Inciso a) - Cálculo de Probabilidades

Para una distribución uniforme continua, la probabilidad se calcula como el área bajo la curva de la función de densidad:

$$P(c < X < d) = \int_c^d f(x)dx = \int_c^d \frac{1}{4}dx = \frac{d-c}{4}$$

#### Parámetros de la distribución

```python
# Parámetros de la distribución uniforme U(a, b)
a = 0  # límite inferior
b = 4  # límite superior

print(f"X ~ U({a}, {b})")
print(f"f(x) = 1/{b-a} para {a} ≤ x ≤ {b}")
```

#### 1. P(X < 2)

```python
# Calcular P(X < 2)
p1_manual = (2 - a) / (b - a)
p1_scipy = uniform.cdf(2, loc=a, scale=b-a)

display(Math(r'P(X < 2) = \int_0^2 \frac{1}{4}dx = \frac{2-0}{4} = 0.5'))
print(f"P(X < 2) = {p1_manual:.4f}")
```

#### 2. P(X ≥ 4)

```python
# Calcular P(X ≥ 4)
p2_manual = 1 - uniform.cdf(4, loc=a, scale=b-a)
p2_scipy = uniform.sf(4, loc=a, scale=b-a)

display(Math(r'P(X \geq 4) = \int_4^4 \frac{1}{4}dx = 0'))
print("Como X ~ U(0, 4), para una variable continua: P(X ≥ 4) = P(X = 4) = 0")
print(f"P(X ≥ 4) = {p2_manual:.4f}")
```

#### 3. P(0 < X < 3)

```python
# Calcular P(0 < X < 3)
p3_manual = (3 - 0) / (b - a)
p3_scipy = uniform.cdf(3, loc=a, scale=b-a) - uniform.cdf(0, loc=a, scale=b-a)

display(Math(r'P(0 < X < 3) = \int_0^3 \frac{1}{4}dx = \frac{3-0}{4} = 0.75'))
print(f"P(0 < X < 3) = {p3_manual:.4f}")
```

#### 4. P(X > 3)

```python
# Calcular P(X > 3)
p4_manual = (b - 3) / (b - a)
p4_scipy = 1 - uniform.cdf(3, loc=a, scale=b-a)

display(Math(r'P(X > 3) = \int_3^4 \frac{1}{4}dx = \frac{4-3}{4} = 0.25'))
print(f"P(X > 3) = {p4_manual:.4f}")
```

---

### Inciso b) - Probabilidades Condicionales

Para calcular probabilidades condicionales, usamos la fórmula:

$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

**Recordatorio:** Para X ~ U(0, 4), tenemos:
- $F(x) = \frac{x}{4}$ para $0 \leq x \leq 4$
- $P(c_1 < X < c_2) = F(c_2) - F(c_1) = \frac{c_2 - c_1}{4}$

#### a) P(X < 3 | X > 2)

```python
# P(X < 3 | X > 2) = P(2 < X < 3) / P(X > 2)
p_interseccion = (3 - 2) / (b - a)
p_x_mayor_2 = (b - 2) / (b - a)
p_cond_a = p_interseccion / p_x_mayor_2

display(Math(r'P(X < 3 \,|\, X > 2) = \frac{P(2 < X < 3)}{P(X > 2)} = \frac{1/4}{1/2} = \frac{1}{2}'))
print(f"P(X < 3 | X > 2) = {p_cond_a:.4f}")
```

#### b) P(X > 1 | X < 3.5)

```python
# P(X > 1 | X < 3.5) = P(1 < X < 3.5) / P(X < 3.5)
p_interseccion_b = (3.5 - 1) / (b - a)
p_x_menor_35 = 3.5 / (b - a)
p_cond_b = p_interseccion_b / p_x_menor_35

display(Math(r'P(X > 1 \,|\, X < 3.5) = \frac{P(1 < X < 3.5)}{P(X < 3.5)} = \frac{2.5/4}{3.5/4} = \frac{2.5}{3.5} = \frac{5}{7}'))
print(f"P(X > 1 | X < 3.5) = {p_cond_b:.4f}")
```

---

### Inciso c) - Función de Distribución Acumulada F(x)

La función de distribución acumulada (FDA) se define como:

$$F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) dt$$

Para la distribución uniforme U(0, 4), debemos considerar tres casos según el valor de x.

#### Deducción de F(x)

```python
# Deducción de F(x) para X ~ U(0, 4)
display(Math(r'F(x) = \begin{cases} 0 & \text{si } x < 0 \\ \frac{x}{4} & \text{si } 0 \leq x \leq 4 \\ 1 & \text{si } x > 4 \end{cases}'))
```

#### Gráfico de F(x)

```python
# Gráfico de F(x)
fig, ax = plt.subplots(figsize=(10, 6))

# Rangos
x1 = np.linspace(-1, 0, 100)
x2 = np.linspace(0, 4, 100)
x3 = np.linspace(4, 5, 100)

# Valores de F(x)
y1 = np.zeros_like(x1)
y2 = x2 / 4
y3 = np.ones_like(x3)

# Graficar
ax.plot(x1, y1, 'b-', linewidth=2, label='F(x) = 0')
ax.plot(x2, y2, 'r-', linewidth=2, label='F(x) = x/4')
ax.plot(x3, y3, 'g-', linewidth=2, label='F(x) = 1')

# Puntos importantes
ax.plot(0, 0, 'ro', markersize=8)
ax.plot(4, 1, 'ro', markersize=8)

# Configuración
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('F(x)', fontsize=12)
ax.set_title('Función de Distribución Acumulada - X ~ U(0, 4)', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left')
ax.set_xlim(-1, 5)
ax.set_ylim(-0.1, 1.2)

plt.tight_layout()
plt.show()
```

# Ejercicio 2

**Sea X una variable aleatoria cuya función de densidad es:**

$$f(x) = \begin{cases} 
\frac{1}{2} - \frac{1}{8}x & \text{si } 0 \leq x \leq 4 \\
0 & \text{en otro caso}
\end{cases}$$

**Calcule:**

a) $P(1 < X \leq 2)$

b) $P(X > 3)$

c) $P(X \leq \frac{1}{2})$

d) $P(X \leq 2 \cup X > 3)$

e) Halle la función de distribución acumulada de $X$.

```python
# Verificar que f(x) integra a 1
x = sp.symbols('x')  # Crear variable simbólica x
f = sp.Rational(1, 2) - sp.Rational(1, 8) * x  # Definir f(x) = 1/2 - x/8

integral_total = sp.integrate(f, (x, 0, 4))  # Calcular ∫₀⁴ f(x)dx
display(Math(r'\int_0^4 \left(\frac{1}{2} - \frac{1}{8}x\right)dx = \left[\frac{x}{2} - \frac{x^2}{16}\right]_0^4 = 2 - 1 = 1'))
print(f"Integral total = {integral_total} ✅")
```

### Inciso a) Calcular P(1 < X ≤ 2)

```python
# P(1 < X ≤ 2) = ∫₁² f(x)dx
prob_a = sp.integrate(f, (x, 1, 2))

display(Math(r'P(1 < X \leq 2) = \int_1^2 \left(\frac{1}{2} - \frac{1}{8}x\right)dx = \left[\frac{x}{2} - \frac{x^2}{16}\right]_1^2'))
display(Math(r'= \left(1 - \frac{1}{4}\right) - \left(\frac{1}{2} - \frac{1}{16}\right) = \frac{3}{4} - \frac{7}{16} = \frac{5}{16}'))
print(f"P(1 < X ≤ 2) = {prob_a} = {float(prob_a):.4f}")
```

### Inciso b) Calcular P(X > 3)

```python
# P(X > 3) = ∫₃⁴ f(x)dx
prob_b = sp.integrate(f, (x, 3, 4))

display(Math(r'P(X > 3) = \int_3^4 \left(\frac{1}{2} - \frac{1}{8}x\right)dx = \left[\frac{x}{2} - \frac{x^2}{16}\right]_3^4'))
display(Math(r'= \left(2 - 1\right) - \left(\frac{3}{2} - \frac{9}{16}\right) = 1 - \frac{15}{16} = \frac{1}{16}'))
print(f"P(X > 3) = {prob_b} = {float(prob_b):.4f}")
```

### Inciso c) Calcular P(X ≤ 1/2)

```python
# P(X ≤ 1/2) = ∫₀^(1/2) f(x)dx
prob_c = sp.integrate(f, (x, 0, sp.Rational(1, 2)))

display(Math(r'P(X \leq \frac{1}{2}) = \int_0^{1/2} \left(\frac{1}{2} - \frac{1}{8}x\right)dx = \left[\frac{x}{2} - \frac{x^2}{16}\right]_0^{1/2}'))
display(Math(r'= \frac{1}{4} - \frac{1}{64} = \frac{16 - 1}{64} = \frac{15}{64}'))
print(f"P(X ≤ 1/2) = {prob_c} = {float(prob_c):.4f}")
```

### Inciso d) Calcular P(X ≤ 2 ∪ X > 3)

```python
# P(X ≤ 2 ∪ X > 3) = P(X ≤ 2) + P(X > 3)
# Como los eventos son mutuamente excluyentes
prob_x_leq_2 = sp.integrate(f, (x, 0, 2))
prob_x_gt_3 = prob_b  # Ya calculado en inciso b

prob_d = prob_x_leq_2 + prob_x_gt_3

display(Math(r'P(X \leq 2 \cup X > 3) = P(X \leq 2) + P(X > 3)'))
display(Math(r'= \int_0^2 f(x)dx + \int_3^4 f(x)dx'))
print(f"P(X ≤ 2) = {prob_x_leq_2} = {float(prob_x_leq_2):.4f}")
print(f"P(X > 3) = {prob_x_gt_3} = {float(prob_x_gt_3):.4f}")
print(f"P(X ≤ 2 ∪ X > 3) = {prob_d} = {float(prob_d):.4f}")
```

### Inciso e) Hallar la función de distribución acumulada F(x)

```python
# Calcular F(x) = ∫₀ˣ f(t)dt para 0 ≤ x ≤ 4
t = sp.symbols('t')
f_t = sp.Rational(1, 2) - sp.Rational(1, 8) * t

F_x = sp.integrate(f_t, (t, 0, x))
F_x_simplificado = sp.simplify(F_x)

display(Math(r'F(x) = \int_0^x \left(\frac{1}{2} - \frac{1}{8}t\right)dt = \left[\frac{t}{2} - \frac{t^2}{16}\right]_0^x = \frac{x}{2} - \frac{x^2}{16}'))
display(Math(r'F(x) = \begin{cases} 0 & \text{si } x < 0 \\ \frac{x}{2} - \frac{x^2}{16} & \text{si } 0 \leq x \leq 4 \\ 1 & \text{si } x > 4 \end{cases}'))
print(f"F(x) = {F_x_simplificado} para 0 ≤ x ≤ 4")
```

```python
# Definir F(x) para graficar
def F_ej2(x_val):
    if x_val < 0:
        return 0
    elif 0 <= x_val <= 4:
        return x_val/2 - x_val**2/16
    else:
        return 1

F_ej2_vectorizada = np.vectorize(F_ej2)

# Gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Rangos
x1 = np.linspace(-1, 0, 100)
x2 = np.linspace(0, 4, 200)
x3 = np.linspace(4, 5, 100)

# Valores de F(x)
y1 = np.zeros_like(x1)
y2 = F_ej2_vectorizada(x2)
y3 = np.ones_like(x3)

# Graficar
ax.plot(x1, y1, 'b-', linewidth=2, label='F(x) = 0')
ax.plot(x2, y2, 'r-', linewidth=2, label=r'F(x) = $\frac{x}{2} - \frac{x^2}{16}$')
ax.plot(x3, y3, 'g-', linewidth=2, label='F(x) = 1')

# Puntos importantes
ax.plot(0, 0, 'ro', markersize=8)
ax.plot(4, 1, 'ro', markersize=8)

# Configuración
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('F(x)', fontsize=12)
ax.set_title(r'Función de Distribución Acumulada - $f(x) = \frac{1}{2} - \frac{1}{8}x$', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left')
ax.set_xlim(-1, 5)
ax.set_ylim(-0.1, 1.2)

plt.tight_layout()
plt.show()
```
