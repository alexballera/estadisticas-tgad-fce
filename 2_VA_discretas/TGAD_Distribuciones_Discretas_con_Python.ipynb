### **DISTRIBUCIÓN BERNOULLI**



Un experimento Bernoulli está asociado a eventos cuyos resultados pueden tomar únicamente dos valores: éxito o fracaso

Entonces, definimos la variable aleatoria X=“cantidad de éxitos en las n repeticiones” que sigue una distribución Bernoulli
\begin{align}
X \sim B(p,1-p)
\end{align}

con p (probabilidad de éxito) y 1-p( probabilidad de fracaso) los parámetros de la distribución


Función de probabilidad:

\begin{align}
P(X=k) = p^{k} (1-p)^{1-k}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)=P(X_i \leq k) = \sum_{i=1}^n p^{k} (1-p)^{1-k}
\end{align}

Valor esperado:
\begin{align}
E(X)=p
\end{align}

Varianza:
\begin{align}
V(X)=p(1-p)
\end{align}

Por ejemplo: sea $X \sim B(0.4)$. Calcule la probabilidad P(X=0) y P(X=1)

```python
ss.bernoulli.pmf(0, 0.4)
```

Se puede verificar el resultado del siguiente modo:

$P(X=0) = 0.4^{0} (1-0.4)^{1-0}$

$P(X=0) = 0.4^{0} (1-0.4)^{1-0}$

$P(X=0) = 0.6$

```python
ss.bernoulli.pmf(1, 0.4)
```

Se puede verificar el resultado del siguiente modo:

$P(X=1) = 0.4^{1} (1-0.4)^{1-1}$

$P(X=1) = 0.4^{1} (1-0.4)^{1-1}$

$P(X=1) = 0.4$

Por ejemplo: sea $X \sim B(0.4)$. Calcule la probabilidad P(X<=1)=F(1)

```python
ss.bernoulli.cdf(1,0.4)
```

Se puede verificar el resultado del siguiente modo:

$F(1)=P(X\leq1) = P(X=0)+P(x=1)$

$P(X=0) = 0.4^{0} (1-0.4)^{1-0}$

$P(X=0) = 0.4^{0} (1-0.4)^{1-0}$

$P(X=0) = 0.6$

$P(X=1) = 0.4^{1} (1-0.4)^{1-1}$

$P(X=1) = 0.4^{1} (1-0.4)^{1-1}$

$P(X=1) = 0.4$

$P(X\leq4) = 0.6 + 0.4$

$P(X\leq4) = 1$

Comportamiento de la distribución Bernoulli:

```python
#Parámetros
p = 0.4 # parametros de forma de la distribución Bernoulli
x = np.arange(0,2,1) #valores de X: 0,1

#Función de Probabilidad
fmp = ss.bernoulli.pmf(x, p)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,10))

#Gráfico de Función de Probabilidad
ax1.bar(x, fmp, color='blue', label='Parámetro: p=0.4', align='center')
ax1.bar_label(ax1.containers[0], label_type='edge')
ax1.set_xticks(x, ['0','1'])
ax1.set_ylim(0,1)
ax1.set_title('Función de Probabilidad')
ax1.set_ylabel('probabilidad')
ax1.set_xlabel('Valores de X')
ax1.legend(fontsize=10)

# Función de Distribución
fmp = ss.bernoulli.cdf(x, p)

#Gráfico de Función de Distribución
ax2.bar(x, fmp, color='blue', label='Parámetro: p=0.4')
ax2.bar_label(ax2.containers[0], label_type='center', color='white')
ax2.set_xticks(x, ['0','1'])
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('Valores de X')
ax2.legend(fontsize=10)

#Parámetros
p_1 = 0.7
x_1 = np.arange(0,2)

#Función de Probabilidad
fmp_1 = ss.bernoulli.pmf(x_1, p_1)

#Gráfico de Función de Probabilidad

ax3.bar(x_1, fmp_1, color='green', label='Parámetro: p=0.7')
ax3.bar_label(ax3.containers[0], label_type='edge')
ax3.set_xticks(x, ['0','1'])
ax3.set_ylim(0,1)
ax3.set_title('Función de Probabilidad')
ax3.set_ylabel('probabilidad')
ax3.set_xlabel('Valores de X')
ax3.legend(fontsize=10)

# Distribución de probabilidad
fmp_1 = ss.bernoulli.cdf(x_1, p_1)

#Gráfico Distribución de probabilidad
ax4.bar(x_1, fmp_1, color='green', label='Parámetro: p=0.7')
ax4.bar_label(ax4.containers[0], label_type='center', color='white')
ax4.set_xticks(x, ['0','1'])
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('Valores de X')
ax4.legend(fontsize=10)

plt.show()
```

### **DISTRIBUCIÓN BINOMIAL**




Un experimento es binomial si satisface las siguientes condiciones:

- El experimento consiste en n pruebas, con n fijo.

- Las pruebas son idénticas y en cada prueba hay sólo dos resultados posibles: éxito y fracaso.

- Las pruebas son independientes, es decir, el resultado de una prueba no influye sobre el de las otras.

- La probabilidad de éxito se mantiene constante en todas las pruebas.

Entonces, definimos la variable aleatoria X=“cantidad de éxitos en las n repeticiones” que sigue una distribución
\begin{align}
X \sim Bi(n,p)
\end{align}

con n (cantidad de ensayos) y p (probabilidad de éxito) los parámetros de la distribución Binomial

Función de probabilidad:

\begin{align}
P(X=k) = {\binom{n}{k}} p^{k} (1-p)^{n-k}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)=P(X_i \leq k) = \sum_{i=1}^n {\binom{n}{k}} p^{k} (1-p)^{n-k}
\end{align}

Valor esperado:
\begin{align}
E(X)=np
\end{align}

Varianza:
\begin{align}
V(X)=np(1-p)
\end{align}

Por ejemplo: sea $X \sim Bi(6,0.4)$. Calcule la probabilidad P(X=4)

```python
ss.binom.pmf(4, 6, 0.4)
```

```python
#si queremso redondear a 4 decimales

round(ss.binom.pmf(4, 6, 0.4),4)
```

Se verifica del siguiente modo:

$P(X=4) = {\binom{6}{4}} 0.4^{4} (1-0.4)^{6-4}$

$P(X=4) = {6!\over4!(6-4)!} 0.4^{4} (1-0.4)^{6-4}$

$P(X=4) \approx 0.1382$

Por ejemplo: sea $X \sim Bi(6,0.4)$. Calcule la probabilidad P(X<=4)

```python
round(ss.binom.cdf(4, 6, 0.4),4)
```

Se verifica del siguiente modo:

$F(4)=P(X\leq4) = P(X=0)+P(x=1)+P(X=2)+P(X=3)+P(X=4)$

$P(X=0) = {\binom{6}{0}} 0.4^{0} (1-0.4)^{6-0}$

$P(X=0) = {6!\over0!(6-0)!} 0.4^{0} (1-0.4)^{6-0}$

$P(X=0) \approx 0.0467$

$P(X=1) = {\binom{6}{1}} 0.4^{1} (1-0.4)^{6-1}$

$P(X=1) = {6!\over1!(6-1)!} 0.4^{1} (1-0.4)^{6-1}$

$P(X=1) \approx 0.1866$

$P(X=2) = {\binom{6}{2}} 0.4^{2} (1-0.4)^{6-2}$

$P(X=2) = {6!\over2!(6-2)!} 0.4^{2} (1-0.4)^{6-2}$

$P(X=2) \approx 0.311$

$P(X=3) = {\binom{6}{3}} 0.4^{3} (1-0.4)^{6-3}$

$P(X=3) = {6!\over3!(6-3)!} 0.4^{3} (1-0.4)^{6-3}$

$P(X=3) \approx 0.2765$

$P(X=4) = {\binom{6}{4}} 0.4^{4} (1-0.4)^{6-4}$

$P(X=4) = {6!\over4!(6-4)!} 0.4^{4} (1-0.4)^{6-4}$

$P(X=4) \approx 0.1382$

$P(X\leq4) = 0.0466 + 0.1866 + 0.311 + 0.2765 + 0.1382$

$P(X\leq4) = 0.9590$

Comportamiento de la distribución binomial:

```python
#Parámetros de la distribución Binomial
n, p = 20, 0.25
n_1, p_1 = 20, 0.7
n_2, p_2 = 20, 0.5

#Valores de X
x =np.arange(0,20,1)
x_1 = np.arange(0,20,1)
x_2 = np.arange(0,40,1)

#Función de Probabilidad
fmp = ss.binom.pmf(x, n, p)
fmp_1 = ss.binom.pmf(x_1, n_1, p_1)
fmp_2 = ss.binom.pmf(x_2, n_2, p_2)

# Distribución de probabilidad
fmpA = ss.binom.cdf(x, n, p)
fmpA_1 = ss.binom.cdf(x_1, n_1, p_1)
fmpA_2 = ss.binom.cdf(x_2, n_2, p_2)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15,15))

#Gráfico de Función de Probabilidad
ax1.bar(x, fmp, color='blue', label='Parámetros: p=0.25  N=20')
ax1.set_xticks(x)
ax1.set_xlim(0,20)
ax1.set_ylim(0,0.25)
ax1.set_title('Función de Probabilidad')
ax1.set_ylabel('probabilidad')
ax1.set_xlabel('valores de X')
ax1.legend(fontsize=10)

#Gráfico de distribución de probabilidad
ax2.bar(x, fmpA, color='blue', label='Parámetros: p=0.25  N=20')
ax2.set_xticks(x)
ax2.set_xlim(0,20)
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('valores de X')
ax2.legend(fontsize=10)

ax3.bar(x_1, fmp_1, color='green', label='Parámetros: p=0.7  N=20')
ax3.set_xticks(x_1)
ax3.set_xlim(0,20)
ax3.set_ylim(0,0.25)
ax3.set_title('Función de Probabilidad')
ax3.set_ylabel('probabilidad')
ax3.set_xlabel('valores de X')
ax3.legend(fontsize=10)

ax4.bar(x_1, fmpA_1, color='green', label='Parámetros: p=0.7  N=20')
ax4.set_xticks(x_1)
ax4.set_xlim(0,20)
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('valores de X')
ax4.legend(fontsize=10)


ax5.bar(x_2, fmp_2, color='red', label='Parámetros: p=0.5  N=20')
ax5.set_xticks(x_2)
ax5.set_xlim(0,20)
ax5.set_ylim(0,0.25)
ax5.set_title('Función de Probabilidad')
ax5.set_ylabel('probabilidad')
ax5.set_xlabel('valores de X')
ax5.legend(fontsize=10)

ax6.bar(x_2, fmpA_2, color='red', label='Parámetros: p=0.5  N=20')
ax6.set_xticks(x_2)
ax6.set_xlim(0,20)
ax6.set_ylim(0,1)
ax6.set_title('Distribución de Probabilidad')
ax6.set_ylabel('F(x)')
ax6.set_xlabel('valores de X')
ax6.legend(fontsize=10)

plt.show()
```

### **DISTRIBUCIÓN GEOMÉTRICA**


Un experimento es geométrico si el interés radica en saber la cantidad de ensayos hasta obtener el primer éxito:

- El experimento consiste en n pruebas, con p fijo.

Entonces, definimos la variable aleatoria X=“cantidad de ensayos dada una probabilidad de éxito” que sigue una distribución Geométrica
\begin{align}
X \sim G(p)
\end{align}

con p (probabilidad de éxito) siendo el parámetro de la distribución

Función de probabilidad:

\begin{align}
P(X=k) =  (1-p)^{k-1} p
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)=P(X_i \leq k) = \sum_{i=1}^n (1-p)^{k-1} p
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {1\over p}
\end{align}

Varianza:
\begin{align}
{V(X)=} {1-p\over p^2}
\end{align}

Por ejemplo: sea $X \sim G(0.4)$. Calcule la probabilidad P(X=4)

```python
round(ss.geom.pmf(4, 0.4),4)
```

Se verifica del siguiente modo:

$P(X=4) = (1-0.4)^{4-1}0.4$

$P(X=4) \approx 0.0864$

Por ejemplo: sea $X \sim G(0.4)$. Calcule la probabilidad P(X<=4)

```python
round(ss.geom.cdf(4, 0.4),4)
```

Se verifica del siguiente modo:

$F(4)=P(X\leq4) = P(x=1)+P(X=2)+P(X=3)+P(X=4)$

$P(X=1) = (1-0.4)^{1-1} 0.4$

$P(X=1) \approx 0.4$

$P(X=2) = (1-0.4)^{2-1} 0.4$

$P(X=2) \approx 0.24$

$P(X=3) = (1-0.4)^{3-1} 0.4$

$P(X=3) \approx 0.144$

$P(X=4) = (1-0.4)^{4-1} 0.4$

$P(X=4) \approx 0.0864$

$P(X\leq4) = 0.4 + 0.24 + 0.144 + 0.0864$

$P(X\leq4) = 0.8704$

Comportamiento de la distribución Geométrica:

```python
#Parámetros de la distribución geométrica
p = 0.2
p_2 = 0.8

#Valores de X
x = np.arange(1,21,1)
x_2 = np.arange(1,21,1)

#Función de Probabilidad
fmp = ss.geom.pmf(x, p)
fmp_2 = ss.geom.pmf(x_2, p_2)

# Distribución de Probabilidad
fmpA = ss.geom.cdf(x, p)
fmpA_2 = ss.geom.cdf(x_2, p_2)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,10))

#Gráfico de Función de probabilidad
ax1.bar(x, fmp,  color='blue', label='Parámetro: p=0.2')
ax1.set_xticks(x)
ax1.set_ylim(0,1)
ax1.set_title('Función de Probabilidad')
ax1.set_ylabel('probabilidad')
ax1.set_xlabel('valores de X')
ax1.legend(fontsize=10)

#Gráfico de distribución de probabilidad
ax2.bar(x, fmpA, color='blue', label='Parámetro: p=0.2')
ax2.set_xticks(x)
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('valores de X')
ax2.legend(fontsize=10)

ax3.bar(x_2, fmp_2, color='red', label='Parámetro: p=0.8')
ax3.set_xticks(x_2)
ax3.set_ylim(0,1)
ax3.set_title('Función de Probabilidad')
ax3.set_ylabel('probabilidad')
ax3.set_xlabel('valores de X')
ax3.legend(fontsize=10)

ax4.bar(x_2, fmpA_2, color='red', label='Parámetro: p=0.8')
ax4.set_xticks(x_2)
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('valores de X')
ax4.legend(fontsize=10)

plt.show()
```

### **DISTRIBUCIÓN HIPERGEOMÉTRICA**

Un experimento es hipergeométrico si satisface las siguientes condiciones:

- Existe uan población N de la cual se toma una muestra de k elementos.

- En N hay M éxitos y N-M fracasos.

- Las pruebas no son independientes, es decir, el resultado de una prueba influye sobre el de las otras.

- La probabilidad de éxito no se mantiene constante en todas las pruebas.

Entonces, definimos la variable aleatoria X=“cantidad de éxitos en las n repeticiones” que sigue una distribución Hipergeométrica
\begin{align}
X \sim H(N,M,n)
\end{align}

con N (cantidad de ensayos), M (cantidad de éxitos) y n (cantidad de selecciones) los parámetros de la distribución


Función de probabilidad:

\begin{align}
{P(X=k) =} {{{\binom{M}{k}} * {\binom{N-M}{n-k}}} \over {\binom{N}{n}}}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)=P(X_i \leq k) = \sum_{i=1}^n {{\binom{M}{k} * {\binom{N-M}{n-k}}} \over {\binom{N}{n}}}
\end{align}

Valor esperado:
\begin{align}
E(X)= n  {M \over N}
\end{align}

Varianza:
\begin{align}
{V(X)=} {n  {M \over N} {({1}- {M \over N})} {({N-n \over N-1})}}
\end{align}

Por ejemplo: sea $X \sim H(7,3,4)$. Calcule la probabilidad P(X=2)

```python
round(ss.hypergeom.pmf(2,7,3,4),4)
```

Se verifica del siguiente modo:

$P(X=2) = {{{\binom{3}{2}} * {\binom{7-3}{4-2}}} \over {\binom{7}{4}}}$

$P(X=2) = {{{3!\over3!(3-2)!} * {4!\over2!(4-2)!}} \over {7!\over4!(7-4)!}}$

$P(X=2) \approx 0.5143$

Por ejemplo: sea $X \sim H(7,3,4)$. Calcule la probabilidad P(X<=2)

```python
round(ss.hypergeom.cdf(2,7,3,4),4)
```

Se verifica del siguiente modo:



\begin{align}
{P(X=k) =} {{{\binom{M}{k}} * {\binom{N-M}{n-k}}} \over {\binom{N}{n}}}
\end{align}

$P(x<=2)= P(X=0) + P(X=1) + P(X=2)$

$P(X=0) = {{{\binom{3}{0}} * {\binom{7-3}{4-0}}} \over {\binom{7}{4}}}$

$P(X=0) = {{{3!\over0!(3-0)!} * {4!\over4!(4-4)!}} \over {7!\over4!(7-4)!}}$

$P(X=0) \approx 0.0285$

$P(X=1) = {{{\binom{3}{1}} * {\binom{7-3}{4-1}}} \over {\binom{7}{4}}}$

$P(X=1) = {{{3!\over1!(3-1)!} * {4!\over3!(4-1)!}} \over {7!\over4!(7-4)!}}$

$P(X=1) \approx 0.3429$

$P(X=2) = {{{\binom{3}{2}} * {\binom{7-3}{4-2}}} \over {\binom{7}{4}}}$

$P(X=2) = {{{3!\over3!(3-2)!} * {4!\over2!(4-2)!}} \over {7!\over4!(7-4)!}}$

$P(X=2) \approx 0.5143$

$P(x<=2)= 0.0286 + 0.3429 + 0.5143 = 0.8857$

### **DISTRIBUCIÓN DE POISSON**



Un experimento es de Poisson si:

- El experimento ocurre a una media (siendo una tasa constante) en el tiempo o espacio.

- Los eventos ocurren de manera independiente del tiempo o espacio

- La variable aleatoria cuenta la cantidad de veces que ocurre un evento considerado exitoso en el intervalo de tiempo o espacio.

Entonces, definimos la variable aleatoria X=“cantidad de casos existosos en las n repeticiones” que sigue una distribución Poisson:
\begin{align}
X \sim P(\lambda)
\end{align}

con ${\lambda}$ (media) siendo el parámetro de la distribución


Función de probabilidad:

\begin{align}
{P(X=k) =}  {{e^{-\lambda}} {\lambda^k} \over k!}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)=P(X_i \leq k) = \sum_{i=1}^k {{e^{-\lambda}} {\lambda^k} \over k!}
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {\lambda}
\end{align}

Varianza:
\begin{align}
{V(X)=} {\lambda}
\end{align}

Por ejemplo: sea $X \sim P(2)$. Calcule la probabilidad P(X=4)

```python
round(ss.poisson.pmf(4, 2),4)
```

Se verifica del siguiente modo:

$P(X=4) = {{e^{-2}} {2^4} \over 4!}$

$P(X=4) \approx 0.0338$

Por ejemplo: sea $X \sim P(2)$. Calcule la probabilidad P(X<=4)

```python
round(ss.poisson.cdf(4, 2),4)
```

Se verifica del siguiente modo:

$F(4)=P(X\leq4) = P(x=1)+P(X=2)+P(X=3)+P(X=4)$

$P(X=0) = {{e^{-2}} {2^0} \over 0!}$

$P(X=0) \approx 0.1353$

$P(X=1) = {{e^{-2}} {2^1} \over 1!}$

$P(X=1) \approx 0.2706$

$P(X=2) = {{e^{-2}} {2^4} \over 4!}$

$P(X=2) \approx 0.2707$

$P(X=3) = {{e^{-2}} {2^3} \over 3!}$

$P(X=3) \approx 0.1805$

$P(X=4) = {{e^{-2}} {2^4} \over 4!}$

$P(X=4) \approx 0.0902$

$P(X\leq4) = 0.1353 + 0.2706 + 0.2707 + 0.1805 + 0.0902$

$P(X\leq4) = 0.9473$
